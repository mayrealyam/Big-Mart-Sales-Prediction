# Feature Scaling

Feature scaling of the divided data

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_train_stand = train_x.copy()
x_test_stand = test_x.copy()
```

```python
num_cols = ["Item_Weight", "Item_Visibility", "Item_MRP", "Outlet_Establishment_Year"]
for i in num_cols:
    scale = StandardScaler().fit(x_train_stand[[i]])
    x_train_stand[i] = scale.transform(x_train_stand[[i]])
    x_test_stand[i] = scale.transform(x_test_stand[[i]])
```

```python
rfr7 = RandomForestRegressor(max_depth=7)
rfr7.fit(x_train_stand, train_y)
pred7 = rfr7.predict(x_test_stand)
print("RMSE for random forest regressor : %.2f" % np.sqrt(np.mean((test_y-pred7)**2)))
```

```python
x_pool = cb.Pool(x_train_stand, train_y)
y_pool = cb.Pool(x_test_stand, test_y)
```

```python
model = cb.CatBoostRegressor(loss_function = "RMSE")
```

```python
grid = {"learning_rate": [0.03, 0.1],
       "depth": [4, 6, 10],
       "l2_leaf_reg": [1, 3, 5],
       "iterations": [50, 100, 150]}
model.grid_search(grid, x_pool)
```
